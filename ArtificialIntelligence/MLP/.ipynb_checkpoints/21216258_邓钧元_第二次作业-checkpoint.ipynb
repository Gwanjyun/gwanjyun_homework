### 1.数据集的定义

```python
class Dataset:
    '''
        不考虑"训练集", "测试集"的划分！
    '''
    def __init__(self):
        self.data, self.labels= datasets.load_iris(return_X_y=True)
        
    def __getitem__(self, index):
        return {'data': self.data[index], 'labels': self.labels[index]}
    
    def __len__(self):
        return len(self.data)
```

### 2.数据加载器的定义

```python
class Dataloader:
    def __init__(self, dataset, batchsize, shuffle = False):
        self.dataset = dataset
        self.batchsize = batchsize
        self.shuffle = shuffle # 是否打乱顺序
        self.len = len(self.dataset) # 数据集长度
        self._shuffle() # 打乱顺序
        self.i = 0 # 一个小标记！
        
    
    def __iter__(self):
        return self
        
    def __next__(self):
        if self.i >= self.len:
            self.i = 0
            self._shuffle() # 打乱顺序
            raise StopIteration()
        # 这东西把dataset[left:right]的东西拿出来
        left = self.i
        right = self.i + self.batchsize
        index = self.index[left:right]
        batch = [self.dataset[i] for i in index]
        data = self.collate_fn(batch)
        self.i = right   
        return data
    
    def next(self):
        return self.__next__()
    
    def collate_fn(self, batch):
        data_dict = {}
        keys = batch[0].keys() # batch中含有的key
        # 把所有的batch弄到一块
        for key in keys:
            value = []
            for data in batch:
                value.append(data[key])
            data_dict.update({key:value})
        # 转为numpy方便矩阵相乘
        for key in keys:
            data_dict[key] = np.array(data_dict[key])
            
        return data_dict
        
    
    def _shuffle(self):
        # 打乱顺序的函数
        if self.shuffle:
            self.index = np.random.choice(self.len, (self.len,), replace = False)
        else:
            self.index = np.arange(self.len)
```

### 3.全连接神经网络的定义

```python
class Linear:
    def __init__(self, in_channels, out_channels):
        '''
            这输入参数不解释了
        '''
        self.W = np.random.randn(in_channels, out_channels)
        self.b = np.random.randn(1, out_channels)
        
        self.grad_W = 0
        self.grad_b = 0
        self._x = 0
        self.grad_x = 0
        # parameters for momentum optimizer
        self.v_W = 0
        self.v_b = 0
        # parameters for Adam optimizer
        self.t = 0
        self.s_W = 0
        self.s_b = 0
        self.r_W = 0
        self.r_b = 0
        
    def forward(self, x):
        self._x = x
        y = x @ self.W + self.b
        return y
    
    def backward(self, grad):
        # 跟老师课件不太一样，这里使用了batch的思想，利用numpy快速运算
        self.grad_W += self._x.T @ grad  
        self.grad_b += np.sum(grad, 0)
        self.grad_x = grad @ (self.W.T)
        return self.grad_x
    
    def step(self, lr, mode, *args):
        if mode == 'SGD':
            self.W = self.W - lr*self.grad_W
            self.b = self.b - lr*self.grad_b
        elif mode == 'Momentum':
            alpha = args[0]
            self.v_W = alpha * self.v_W - lr*self.grad_W
            self.v_b = alpha * self.v_b - lr*self.grad_b
            
            self.W = self.W + self.v_W
            self.b = self.b + self.v_b
            
        elif mode == 'Adam':
            delta = 1e-12
            rho1 = args[0] # 0.9
            rho2 = args[1] # 0.999
            self.t = self.t + 1
            
            self.s_W = rho1 * self.s_W + (1 - rho1) * self.grad_W
            self.s_b = rho1 * self.s_b + (1 - rho1) * self.grad_b
            self.r_W = rho2 * self.r_W + (1 - rho2) * (self.grad_W * self.grad_W)
            self.r_b = rho2 * self.r_b + (1 - rho2) * (self.grad_b * self.grad_b)
            
            self.s_W_hat = self.s_W/(1 - rho1**self.t)
            self.s_b_hat = self.s_b/(1 - rho1**self.t)
            self.r_W_hat = self.r_W/(1 - rho2**self.t)
            self.r_b_hat = self.r_b/(1 - rho2**self.t)
            
            self.theta_W_hat = -lr * self.s_W_hat/(self.r_W_hat**0.5 + delta)
            self.theta_b_hat = -lr * self.s_b_hat/(self.r_b_hat**0.5 + delta)
            
            self.W = self.W + self.theta_W_hat
            self.b = self.b + self.theta_b_hat
            
            
    
    def zero_grad(self):
        self.grad_W = 0
        self.grad_b = 0
        self.grad_x = 0
    
    def __call__(self, *args):
        return self.forward(*args)
    
```

### 4.激活函数 LeakyReLU, Sigmoid

```python
class LeakyReLU:
    def __init__(self, ratio = 0):
        '''
            这玩意用了还会有可能不收敛的，放在这占空间而已
        '''
        self._x = 0
        self.grad_x = 0
        self.ratio = ratio
    
    def forward(self, x):
        self._x = x
        self.mask = (x>0)*1 + self.ratio * (x < 0)
        y = x*self.mask
        return y
    
    def backward(self, grad):
        self.grad_x = grad * self.mask
        return self.grad_x
    
    def step(self, *args):
        pass
    
        
    def zero_grad(self):
        self.grad_x = 0
    
    def __call__(self, *args):
        return self.forward(*args)
```

```python
class Sigmoid:
    def __init__(self):
        '''
            Sigmoid
        '''
        self._x = 0
        self.grad_x = 0
    
    def forward(self, x):
        y = 1/(1 + np.exp(-x))
        self._x = y
        return y
    
    def backward(self, grad):
        self.grad_x = grad * self._x*(1 - self._x)
        return self.grad_x
    
    def step(self, *args):
        pass
    
        
    def zero_grad(self):
        self.grad_x = 0
    
    def __call__(self, *args):
        return self.forward(*args)
```

### 5.Loss Function: MSELoss, CrossEntropyLoss, SoftmaxCrossEntropyLoss

```python
class MSELoss:
    def __init__(self, dim = -1):
        self.dim = dim
        self._x = 0
        self.grad_x = 0
    
    def forward(self, x, target):
        self._x = x
        t = np.zeros_like(self._x)
        for i in range(t.shape[0]):
            index = target[i]
            t[i][index] = 1
        
        self.grad_x = (x - t)
        loss = np.sum((x - t)**2)
        return loss
    
    def backward(self, *args):
        return self.grad_x
    
    def step(self, *args):
        pass
    
        
    def zero_grad(self):
        self.grad_x = 0
        
    
    def __call__(self, *args):
        return self.forward(*args)
```

```python
class Softmax:
    def __init__(self, dim = -1):
        self.dim = dim
        self._x = 0
        self.grad_x = 0
    
    def forward(self, x):
        x = x - np.max(x, -1, keepdims=True)
        x_exp = np.exp(x)
        den = np.sum(x_exp, -1, keepdims = True)
        y = x_exp/den
        self._x = y
        return y
    
    def backward(self, grad):
        self.grad_x = grad * self._x*(1 - self._x)
        return self.grad_x
    
    def step(self, *args):
        pass
    
        
    def zero_grad(self):
        self.grad_x = 0
        
    
    def __call__(self, *args):
        return self.forward(*args)
```

```python
class CrossEntropyLoss:
    def __init__(self):
        self._x = 0
        self.grad_x = 0
        
    def forward(self, x, target):
        t = np.zeros_like(x)
        loss = 0
        for i in range(t.shape[0]):
            index = target[i]
            t[i][index] = 1
            loss += -np.log(x[i][index])
            
        
        loss = loss
        self.grad_x = -(t/(x + 1e-12) + (t-1)/(1-x + 1e-12))
        self.grad_x = self.grad_x
        
        
        return loss
        
    def backward(self, *args):
        return self.grad_x
    
    def step(self, *args):
        pass
        
    def zero_grad(self):
        self.grad_x = 0
    
    def __call__(self, *args):
        return self.forward(*args)
```

```python
class SoftmaxCrossEntropyLoss:
    def __init__(self, dim = -1):
        self.dim = dim
        self._x = 0
        self.grad_x = 0
    
    def forward(self, x, target):
        x = x - np.max(x, -1, keepdims=True)
        x_exp = np.exp(x)
        den = np.sum(x_exp, -1, keepdims = True)
        y = x_exp/den
        self._x = y
        
        t = np.zeros_like(self._x)
        for i in range(t.shape[0]):
            index = target[i]
            t[i][index] = 1
        self.grad_x = -(t*(1 - y) + (t - 1)*y)
        self.grad_x = self.grad_x
        loss = np.sum(-t*np.log(y+1e-12))
        return loss
    
    def backward(self, *args):
        return self.grad_x
    
    def step(self, *args):
        pass
    
        
    def zero_grad(self):
        self.grad_x = 0
        
    
    def __call__(self, *args):
        return self.forward(*args)
```

### 6.网络的搭建

```python
class Model:
    def __init__(self, in_channels, out_channels, layers = [8,16,8,4]):
        '''
            搭建网络，放在self.layers里面
        '''
        self.layers = []
        in_planes = in_channels
        for layer in layers[:-1]:
            out_planes = layer
            self.layers.append(Linear(in_planes, out_planes))
#             self.layers.append(LeakyReLU()) # 这玩意在优化器为Adam的时候才收敛！！！ 尽量不用了！！
            self.layers.append(Sigmoid())
            in_planes = out_planes
        self.layers.append(Linear(in_planes, out_channels))
        self.layers.append(Softmax())
        
    def forward(self, x):
        for layer in self.layers:
            x = layer(x)
        return x
    
    def backward(self, grad):
        for layer in self.layers[::-1]:
            grad = layer.backward(grad)
        return grad
    
    def step(self, lr, mode, *args):
        for layer in self.layers[::-1]:
            layer.step(lr, mode, *args)
        
    def zero_grad(self):
        for layer in self.layers[::-1]:
            layer.zero_grad()
    
    def __call__(self, *args):
        return self.forward(*args)
    
    def train(self):
        self.mode = 'train'
```

### 7.初始化数据集和数据加载器

```python
dataset = Dataset()
dataloader = Dataloader(dataset, 32, True)
```

### 8.模型的一些重要参数

```python
m = Model(4,3,[8,16,8,4])
m1 = copy.deepcopy(m)
m2 = copy.deepcopy(m)
loss_func = CrossEntropyLoss()
loss_func1 = CrossEntropyLoss()
loss_func2 = CrossEntropyLoss()
```

### 9.模型训练

```python
epoch_iter = trange(3000)

loss_array = []
loss_array1 = []
loss_array2 = []

acc_array = []
acc_array1 = []
acc_array2 = []

for i in epoch_iter:
    num = 0
    num1 = 0
    num2 = 0
    
    loss_all = 0
    loss_all1 = 0
    loss_all2 = 0
    for x in dataloader:
        y = m(x['data'])
        y1 = m1(x['data'])
        y2 = m2(x['data'])
        
        loss = loss_func(y, x['labels'])
        loss1 = loss_func1(y1, x['labels'])
        loss2 = loss_func2(y2, x['labels'])
        
        loss_all += loss
        loss_all1 += loss1
        loss_all2 += loss2

        m.zero_grad()
        m1.zero_grad()
        m2.zero_grad()
        
        grad = loss_func.backward()
        grad1 = loss_func1.backward()
        grad2 = loss_func2.backward()
        
        m.backward(grad)
        m1.backward(grad1)
        m2.backward(grad2)
        
        m.step(0.001, mode = 'SGD')
        m1.step(0.001, 'Momentum', 0.9)
        m2.step(0.001, 'Adam', 0.9, 0.999)
        
        num += np.sum(x['labels'] == np.argmax(y,1))
        num1 += np.sum(x['labels'] == np.argmax(y1,1))
        num2 += np.sum(x['labels'] == np.argmax(y2,1))
        
    acc = num/len(dataset)
    acc1 = num1/len(dataset)
    acc2 = num2/len(dataset)
    
    loss_array.append(loss_all)
    loss_array1.append(loss_all1)
    loss_array2.append(loss_all2)
    
    acc_array.append(acc)
    acc_array1.append(acc1)
    acc_array2.append(acc2)
    
    
    
    epoch_iter.set_postfix(loss = ' {loss_all:.2f} - {loss_all1:.2f} - {loss_all2:.2f} '.format(loss_all = loss_all, loss_all1 = loss_all1, loss_all2 = loss_all2), 
                           acc = ' {acc:.2f}% - {acc1:.2f}% - {acc2:.2f}% '.format(acc = acc*100, acc1 = acc1*100, acc2 = acc2*100))
```

### 10.画图

```python
plt.figure(figsize = (12,6))
plt.plot(np.arange(len(loss_array)), loss_array, label = 'SGD')
plt.plot(np.arange(len(loss_array1)), loss_array1, label = 'Momentum')
plt.plot(np.arange(len(loss_array2)), loss_array2, label = 'Adam')
plt.title('Loss Curve', fontsize = 12)
plt.xlabel('Epoch', fontsize = 12)
plt.ylabel('Loss', fontsize = 12)
plt.legend(fontsize = 12)
plt.tick_params(labelsize = 12)
plt.grid()
plt.show()
```

## 零.构建Beale 公式及其导数

### 0.导入相关模块

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as plt_cl
```

### 1.构建相关函数

```python
# 定义beale公式
def beale(x1,x2):
    return (1.5-x1+x1*x2)**2+(2.25-x1+x1*x2**2)**2+(2.625-x1+x1*x2**3)**2
# 定义beale公式的偏导函数
def dbeale_dx(x1, x2):
    dfdx1 = 2*(1.5-x1+x1*x2)*(x2-1)+2*(2.25-x1+x1*x2**2)*(x2**2-1)+2*(2.625-x1+x1*x2**3)*(x2**3-1)
    dfdx2 = 2*(1.5-x1+x1*x2)*x1+2*(2.25-x1+x1*x2**2)*(2*x1*x2)+2*(2.625-x1+x1*x2**3)*(3*x1*x2**2)
    return dfdx1, dfdx2 
```

### 2.可视化beale方程

```python
# 定义画图函数
def gd_plot(x_traj):
    plt.rcParams['figure.figsize'] = [6, 6]
    plt.contour(X1, X2, Y, levels=np.logspace(0, 6, 30),
                norm=plt_cl.LogNorm(), cmap=plt.cm.jet)
    plt.title('2D Contour Plot of Beale function(Momentum)')
    plt.xlabel('$x_1$')
    plt.ylabel('$x_2$')
    plt.axis('equal')
    plt.plot(3, 0.5, 'k*', markersize=10)
    if x_traj is not None:
        x_traj = np.array(x_traj)
        plt.plot(x_traj[:, 0], x_traj[:, 1], 'k-')
    plt.show()

step_x1, step_x2 = 0.2, 0.2
X1, X2 = np.meshgrid(np.arange(-5, 5 + step_x1, step_x1),
                     np.arange(-5, 5 + step_x2, step_x2))
Y = beale(X1, X2)
print("目标结果 (x_1, x_2) = (3, 0.5)")
gd_plot(None)
```

## 一.编写不使用优化器优化Beale 公式

### 0.构建无优化器函数

```python
# 定义无优化器函数
def gd_no(df_dx, x0, conf_para=None):
    if conf_para is None:
        conf_para = {}
    conf_para.setdefault('n_iter', 1000)  # 迭代次数
    conf_para.setdefault('learning_rate', 0.001)  # 设置学习率
    x_traj = []
    x_traj.append(x0)
    v = np.zeros_like(x0)
    # 没有迭代更新的操作，所以，坐标没有变化
    for iter in range(1, conf_para['n_iter'] + 1):
        x_traj.append(x_traj[-1])
    return x_traj 
```

### 1.优化beale方程

```python
# 初始化坐标
x0 = np.array([1.0, 1.5])
conf_para_no = {'n_iter': 2000, 'learning_rate': 0.005}
# 调用函数进行更新
x_traj_no = gd_no(dbeale_dx, x0, conf_para_no)
print("无优化器求得极值点 (x_1, x_2) = (%s, %s)" % (x_traj_no[-1][0], x_traj_no[-1][1]))
gd_plot(x_traj_no) 
```

## 二.编写SGD优化器并优化Beale 公式

### 0.构建SGD优化器

```python
def gd_sgd(df_dx, x0, conf_para=None):
    if conf_para is None:
        conf_para = {}
    conf_para.setdefault('n_iter', 1000)  # 迭代次数
    conf_para.setdefault('learning_rate', 0.001)  # 设置学习率
    x_traj = []
    x_traj.append(x0)
    v = np.zeros_like(x0)
    # 利用梯度值对坐标进行更新
    for iter in range(1, conf_para['n_iter'] + 1):
        dfdx = np.array(df_dx(x_traj[-1][0], x_traj[-1][1]))
        v = - conf_para['learning_rate'] * dfdx
        x_traj.append(x_traj[-1] + v)
    return x_traj 
```

### 1.优化beale方程

```python
x0 = np.array([1.0, 1.5])
conf_para_sgd = {'n_iter': 2000, 'learning_rate': 0.005}
x_traj_sgd = gd_sgd(dbeale_dx, x0, conf_para_sgd)
print("SGD求得极值点 (x_1, x_2) = (%s, %s)" % (x_traj_sgd[-1][0], x_traj_sgd[-1][1]))
gd_plot(x_traj_sgd) 
```

## 二.编写动量优化器并优化Beale 公式

### 0.构建Momentum优化器

```python
def gd_momentum(df_dx, x0, conf_para=None):
    if conf_para is None:
        conf_para = {}
    conf_para.setdefault('n_iter', 1000)  # 迭代次数
    conf_para.setdefault('learning_rate', 0.001)  # 设置学习率
    conf_para.setdefault('momentum', 0.9)  # 设置动量参数
    x_traj = []
    x_traj.append(x0)
    v = np.zeros_like(x0)
    # 套用动量优化器公式，对坐标值进行更新
    for iter in range(1, conf_para['n_iter'] + 1):
        dfdx = np.array(df_dx(x_traj[-1][0], x_traj[-1][1]))
        v = conf_para['momentum'] * v - conf_para['learning_rate'] * dfdx
        x_traj.append(x_traj[-1] + v)
    return x_traj 
```

### 1.优化beale方程

```python
x0 = np.array([1.0, 1.5])
conf_para_momentum = {'n_iter': 500, 'learning_rate': 0.005}
x_traj_momentum = gd_momentum(dbeale_dx, x0, conf_para_momentum)
print("Momentum求得极值点 (x_1, x_2) = (%s, %s)" % (x_traj_momentum[-1][0], x_traj_momentum[-1][1]))
gd_plot(x_traj_momentum) 
```

## 三、编写自适应优化器并优化Beale 公式

### 0.编写adagrad优化器

```python

def gd_adagrad(df_dx, x0, conf_para=None):
    if conf_para is None:
        conf_para = {}
    conf_para.setdefault('n_iter', 1000)  # 迭代次数
    conf_para.setdefault('learning_rate', 0.001)  # 学习率
    conf_para.setdefault('epsilon', 1e-7)
    x_traj = []
    x_traj.append(x0)
    r = np.zeros_like(x0)
   # 套用adagrad优化器公式，对参数进行更新
    for iter in range(1, conf_para['n_iter'] + 1):
        dfdx = np.array(df_dx(x_traj[-1][0], x_traj[-1][1]))
        r += dfdx ** 2
        x_traj.append(x_traj[-1] - conf_para['learning_rate'] / (np.sqrt(r) + conf_para['epsilon']) * dfdx)
    return x_traj 
```

### 1.优化beale方程

```python
x0 = np.array([1.0, 1.5])
conf_para_adag = {'n_iter': 500, 'learning_rate': 2}
x_traj_adag = gd_adagrad(dbeale_dx, x0, conf_para_adag)
print("Adagrad求得极值点 (x_1, x_2) = (%s, %s)" % (x_traj_adag[-1][0], x_traj_adag[-1][1]))
gd_plot(x_traj_adag) 
```

## 四.导入模块、读取鸢尾花数据集并预处理、构建神经网络

```python
import csv
import os
import time
import numpy as np
from easydict import EasyDict as edict
from matplotlib import pyplot as plt

import mindspore
from mindspore import nn
from mindspore import context
from mindspore import dataset
from mindspore.train.callback import TimeMonitor, LossMonitor
from mindspore import Tensor
from mindspore.train import Model
from mindspore.train.callback import ModelCheckpoint, CheckpointConfig

context.set_context(mode=context.GRAPH_MODE, device_target="CPU")
# 设定运行模式为静态图模式，并且运行设备为昇腾芯片

# 变量定义
cfg = edict({
    'data_size': 150,
    'train_size': 120,      # 训练集大小
    'test_size': 30 ,       # 测试集大小
    'feature_number': 4,       # 输入特征数
    'num_class': 3,     # 分类类别
    'batch_size': 30,
    'data_dir':    'iris.csv',                     
    'save_checkpoint_steps': 5,                 # 多少步保存一次模型
    'keep_checkpoint_max': 1,                      # 最多保存多少个模型
    'out_dir_no_opt':   './model_iris/no_opt',          # 保存模型路径，无优化器模型
    'out_dir_sgd':   './model_iris/sgd',          # 保存模型路径,SGD优化器模型
    'out_dir_momentum':   './model_iris/momentum',          # 保存模型路径，momentum模型
    'out_dir_adam':   './model_iris/adam',          # 保存模型路径，adam优化器模型
    'output_prefix': "checkpoint_fashion_forward"     # 保存模型文件名
})
```

### 1.读取数据并预处理

```python
# 读取鸢尾花数据集，本数据集共有150个带标签的数据
with open(cfg.data_dir) as csv_file:
    data = list(csv.reader(csv_file, delimiter=','))

# 共150条数据，将数据集的4个属性作为自变量X。将数据集的3个类别映射为{0, 1，2}，作为因变量Y
label_map = {'Iris-setosa': 0,'Iris-versicolor': 1,'Iris-virginica':2 }
# 分别获取数据中的特征值X和标签值Y
X = np.array([[float(x) for x in s[:-1]] for s in data[:cfg.data_size]], np.float32)
Y = np.array([label_map[s[-1]] for s in data[:cfg.data_size]], np.int32)

# 将数据集分为训练集120条，测试集30条。
train_idx = np.random.choice(cfg.data_size, cfg.train_size, replace=False)
test_idx = np.array(list(set(range(cfg.data_size)) - set(train_idx)))
X_train, Y_train = X[train_idx], Y[train_idx]
X_test, Y_test = X[test_idx], Y[test_idx]

# 使用MindSpore GeneratorDataset接口将numpy.ndarray类型的数据转换为Dataset
def gen_data(X_train, Y_train, epoch_size):
    # 生成训练集
    XY_train = list(zip(X_train, Y_train))
    ds_train = dataset.GeneratorDataset(XY_train, ['x', 'y'])
    # 设定数据集大小
    # 打乱操作并设定batchsize
    ds_train = ds_train.shuffle(buffer_size=cfg.train_size).batch(cfg.batch_size, drop_remainder=True)
    # 生成测试集
    XY_test = list(zip(X_test, Y_test))
    ds_test = dataset.GeneratorDataset(XY_test, ['x', 'y'])
    # 设定数据集大小
    # 打乱操作并设定batchsize
    ds_test = ds_test.shuffle(buffer_size=cfg.test_size).batch(cfg.test_size, drop_remainder=True)
    return ds_train, ds_test  
```

### 2.定义训练、评估、测试函数

```python
# 训练函数
def train(network, net_opt, ds_train, prefix, directory, print_times):
    # 定义网络损失函数
    net_loss = nn.SoftmaxCrossEntropyWithLogits(sparse=True, reduction="mean")
    # 定义模型
    model = Model(network, loss_fn=net_loss, optimizer=net_opt, metrics={"acc"})
    # 定义损失值指标
    loss_cb = LossMonitor(per_print_times=int(cfg.train_size / cfg.batch_size))
    # 设置checkpoint
    config_ck = CheckpointConfig(save_checkpoint_steps=cfg.save_checkpoint_steps,
                                 keep_checkpoint_max=cfg.keep_checkpoint_max)
    ckpoint_cb = ModelCheckpoint(prefix=prefix, directory=directory, config=config_ck)
    print("============== Starting Training ==============")
    # 训练模型
    model.train(epoch_size, ds_train, callbacks=[ckpoint_cb, loss_cb], dataset_sink_mode=False)
    return model

class_names=['setosa', 'versicolor', 'virginica']
# 评估预测函数
def eval_predict(model, ds_test):
    # 使用测试集评估模型，打印总体准确率
    metric = model.eval(ds_test)
    print(metric)
    # 预测
    test_ = ds_test.create_dict_iterator().__next__()
    test = Tensor(test_['x'], mindspore.float32)
    predictions = model.predict(test)
    predictions = predictions.asnumpy()
    true_label = test_['y'].asnumpy()
    for i in range(10):
        p_np = predictions[i, :]
        pre_label = np.argmax(p_np)
        print('第' + str(i) + '个sample预测结果：', class_names[pre_label], '   真实结果：', class_names[true_label[i]]) 
```

## 五.无优化器模型训练并预测

```python
epoch_size = 20   # 训练20次
print('------------------无优化器--------------------------')
# 数据
ds_train, ds_test = gen_data(X_train, Y_train, epoch_size)  # 生成训练集和测试集
# 定义网络并训练
network = nn.Dense(cfg.feature_number, cfg.num_class)  # 定义一个全连接网络层，输入特征为4，输出类别为3
model = train(network, None, ds_train, "checkpoint_no_opt", cfg.out_dir_no_opt, print_times=4)  # 用训练集训练网络，设置网络结构，模型名称，保存路径, print_times
# 评估预测
eval_predict(model, ds_test)  # 用测试集进行预测 
```

## 六.SGD优化器模型训练并预测

```python
# ---------------------------------------------------SGD-------------------------------------
epoch_size = 300  # 训练300次
lr = 0.01
print('-------------------SGD优化器-----------------------')
# 数据
ds_train, ds_test = gen_data(X_train, Y_train, epoch_size) # 生成训练集和测试集
# 定义网络并训练、测试、预测
network = nn.Dense(cfg.feature_number, cfg.num_class)  # 定义一个全连接网络层，输入特征为4，输出类别为3
net_opt = nn.SGD(network.trainable_params(), lr)  # 用SGD优化器进行优化 
model = train(network, net_opt, ds_train, "checkpoint_sgd", cfg.out_dir_sgd, 40)  # 用训练集训练网络，设置网络结构，优化器，模型名称，保存路径, print_times
# 评估预测
eval_predict(model, ds_test) # 用测试集进行预测       
```

## 七.Momentum优化器模型训练并预测

```python
epoch_size = 20  # 训练20次
lr = 0.01  # 学习率为0.01
print('-------------------Momentum优化器-----------------------')
# 数据
ds_train, ds_test = gen_data(X_train, Y_train, epoch_size)  # 生成训练集和测试集
# 定义网络并训练
network = nn.Dense(cfg.feature_number, cfg.num_class) # 定义一个全连接网络层，输入特征为4，输出类别为3
net_opt = nn.Momentum(network.trainable_params(), lr, 0.9)   # 用 momentum 优化器进行优化，学习率为0.01，动量大小为0.9
model = train(network, net_opt, ds_train, "checkpoint_momentum", cfg.out_dir_momentum, 4)  # 用训练集训练网络，设置网络结构，优化器，模型名称，保存路径, print_times
# 评估预测
eval_predict(model, ds_test)  # 用测试集进行预测
```

## 八.Adam优化器模型训练并预测

```python
epoch_size = 15  # 训练15次
lr = 0.1  # 学习率为0.1, 动态学习率
print('------------------Adam优化器--------------------------')
# 数据
ds_train, ds_test = gen_data(X_train, Y_train, epoch_size)  # 生成训练集和测试集
# 定义网络并训练
network = nn.Dense(cfg.feature_number, cfg.num_class)  # 定义一个全连接网络层，输入特征为4，输出类别为3
net_opt = nn.Adam(network.trainable_params(), learning_rate=lr)  # 用 Adam 优化器进行优化，学习率为0.1
model = train(network, net_opt, ds_train, "checkpoint_adam", cfg.out_dir_adam, 4)  # 用训练集训练网络，设置网络结构，优化器，模型名称，保存路径, print_times
# 评估预测
eval_predict(model, ds_test) 
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
```

## 0.数据集定义

```python
class Dataset(torch.utils.data.Dataset):
    '''
        不考虑"训练集", "测试集"的划分！
    '''
    def __init__(self):
        super().__init__()
        self.data, self.labels= datasets.load_iris(return_X_y=True)
        
    def __getitem__(self, index):
        return {'data': self.data[index].astype(np.float32), 'labels': self.labels[index].astype(np.int64)}
    
    def __len__(self):
        return len(self.data)
```

## 1.实例化数据集和数据加载器

```python
dataset = Dataset()
dataloader = torch.utils.data.DataLoader(dataset, 32, shuffle = True)
```

## 2.构建模型

```python
model = nn.Sequential(
    nn.Linear(4,8),
    nn.ReLU(),
    nn.Linear(8,16),
    nn.ReLU(),
    nn.Linear(16,8),
    nn.ReLU(),
    nn.Linear(8,4),
    nn.ReLU(),
    nn.Linear(4,3),
#     nn.Softmax(-1)
)
model1 = copy.deepcopy(model)
model2 = copy.deepcopy(model)
```

## 3.损失函数和优化器

```python
loss_func = nn.CrossEntropyLoss(reduction = 'sum')
optimizer = optim.SGD(model.parameters(), 0.001)
optimizer1 = optim.SGD(model1.parameters(), 0.001, momentum = 0.9)
optimizer2 = optim.Adam(model2.parameters(), 0.001, betas = (0.9,0.999))
```

## 4.模型训练

```python
epoch_iter = trange(3000)

loss_array = []
loss_array1 = []
loss_array2 = []

acc_array = []
acc_array1 = []
acc_array2 = []

for i in epoch_iter:
    num = 0
    num1 = 0
    num2 = 0
    
    loss_all = 0
    loss_all1 = 0
    loss_all2 = 0
    for x in dataloader:
        y = model(x['data'])
        y1 = model1(x['data'])
        y2 = model2(x['data'])
        
        loss = loss_func(y, x['labels'])
        loss1 = loss_func(y1, x['labels'])
        loss2 = loss_func(y2, x['labels'])

        optimizer.zero_grad()
        optimizer1.zero_grad()
        optimizer2.zero_grad()
        
        loss.backward()
        loss1.backward()
        loss2.backward()
        
        
        optimizer.step()
        optimizer1.step()
        optimizer2.step()
        
        loss_all += float(loss)
        loss_all1 += float(loss1)
        loss_all2 += float(loss2)
        num += torch.sum(x['labels'] == torch.argmax(y,1))
        num1 += torch.sum(x['labels'] == torch.argmax(y1,1))
        num2 += torch.sum(x['labels'] == torch.argmax(y2,1))
        
    acc = num/len(dataset)
    acc1 = num1/len(dataset)
    acc2 = num2/len(dataset)
    
    loss_array.append(loss_all)
    loss_array1.append(loss_all1)
    loss_array2.append(loss_all2)
    
    acc_array.append(acc)
    acc_array1.append(acc1)
    acc_array2.append(acc2)
    
    
    
    epoch_iter.set_postfix(loss = ' {loss_all:.2f} - {loss_all1:.2f} - {loss_all2:.2f} '.format(loss_all = loss_all, loss_all1 = loss_all1, loss_all2 = loss_all2), 
                           acc = ' {acc:.2f}% - {acc1:.2f}% - {acc2:.2f}% '.format(acc = acc*100, acc1 = acc1*100, acc2 = acc2*100))
```
